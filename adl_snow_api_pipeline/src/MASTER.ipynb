```python

##Reading lakehouse tables

adl_tab =    spark.table("zeyodb.adl_tab")
adl_tab.display()

snow_tab=    spark.table("zeyodb.snow_tab")
snow_tab.display()

api_tab =    spark.table("zeyodb.api_tab")
api_tab.display()


##Joining the tables

joindf = (
        adl_tab
        .join(snow_tab,["username"],"left")
        .join(api_tab,["username"],"left")
)


joindf.display()


##Writing the result to another lakehouse table


joindf.write.mode("overwrite").saveAsTable("zeyodb.master_tab")
```
