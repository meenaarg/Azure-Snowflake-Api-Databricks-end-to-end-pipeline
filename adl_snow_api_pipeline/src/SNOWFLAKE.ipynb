```python
##Reading data from snowflake

sdf = (
    spark.read.format("snowflake")
    .option("sfURL", "https://tajfnec-hu24006.snowflakecomputing.com")
    .option("sfAccount", "tajfnec")
    .option("sfUser", "thilagazeyo")
    .option("sfPassword", "Zeyobron@usa908" )   # this variable holds the password
    .option("sfDatabase", "zeyodb")
    .option("sfSchema", "zeyoschema")
    .option("sfRole", "ACCOUNTADMIN")
    .option("sfWarehouse", "COMPUTE_WH")
    .option("dbtable", "srctab")
    .load()
)


sdf.show()

```

```python
##Transformation - Every IP used for every username

from pyspark.sql.functions import *

ipdf = sdf.groupBy("USERNAME")\
    .agg(count("IP"))\
        .withColumnRenamed("count(IP)","IPs")


ipdf.show(5)
```

```python
##Write the data to the lakehouse table 

ipdf.write.mode("overwrite").saveAsTable("zeyodb.snow_tab")

```
